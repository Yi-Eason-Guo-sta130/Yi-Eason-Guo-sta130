```python
# Question 2
# ChatBot version 1

import pandas as pd
import scipy.stats as stats

# Load the penguins dataset
pingees = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Drop rows with missing values in 'flipper_length_mm'
pingees_clean = pingees.dropna(subset=['flipper_length_mm'])

# Calculate statistics
mean_flipper_length = pingees_clean['flipper_length_mm'].mean()
median_flipper_length = pingees_clean['flipper_length_mm'].median()
mode_flipper_length = pingees_clean['flipper_length_mm'].mode()[0]  # Mode returns a Series
std_flipper_length = pingees_clean['flipper_length_mm'].std(ddof=1)  # Sample standard deviation
variance_flipper_length = pingees_clean['flipper_length_mm'].var(ddof=1)  # Sample variance
range_flipper_length = pingees_clean['flipper_length_mm'].max() - pingees_clean['flipper_length_mm'].min()

# Display the results
print(f"Mean: {mean_flipper_length:.2f}")
print(f"Median: {median_flipper_length:.2f}")
print(f"Mode: {mode_flipper_length:.2f}")
print(f"Range: {range_flipper_length:.2f}")
print(f"Variance: {variance_flipper_length:.2f}")
print(f"Standard Deviation: {std_flipper_length:.2f}")

# Probability distribution (using Kernel Density Estimation)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.kdeplot(pingees_clean['flipper_length_mm'], fill=True, color='skyblue', bw_adjust=0.5)
plt.title('Probability Density Function of Flipper Lengths')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Density')
plt.axvline(mean_flipper_length, color='blue', linestyle='--', label='Mean')
plt.axvline(median_flipper_length, color='green', linestyle='--', label='Median')
plt.axvline(mode_flipper_length, color='red', linestyle='--', label='Mode')
plt.legend()
plt.show()
```

```python
#Question 2
#Chatbot version 2


import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the penguins dataset
pingees = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Drop rows with missing values in 'flipper_length_mm'
pingees_clean = pingees.dropna(subset=['flipper_length_mm'])

# Calculate mean, median, mode, standard deviation, variance, and range for flipper_length_mm
mean_flipper_length = pingees_clean['flipper_length_mm'].mean()
median_flipper_length = pingees_clean['flipper_length_mm'].median()
mode_flipper_length = pingees_clean['flipper_length_mm'].mode()[0]  # Mode returns a Series
std_flipper_length = pingees_clean['flipper_length_mm'].std(ddof=1)

# Create the histogram of flipper lengths
fig = px.histogram(
    pingees_clean, 
    x="flipper_length_mm", 
    nbins=20, 
    title="Histogram of Penguin Flipper Lengths with Annotations",
    labels={"flipper_length_mm": "Flipper Length (mm)"}
)

# Add lines for mean, median, mode
fig.add_vline(x=mean_flipper_length, line_dash="dash", line_color="blue")
fig.add_vline(x=median_flipper_length, line_dash="dash", line_color="green")
fig.add_vline(x=mode_flipper_length, line_dash="dash", line_color="red")

# Add a rectangle to show 1 standard deviation range
fig.add_vrect(x0=mean_flipper_length - std_flipper_length, x1=mean_flipper_length + std_flipper_length, 
              fillcolor="lightblue", opacity=0.2, line_width=0)

# Add text annotations with the values of mean, median, mode, and standard deviation
fig.add_annotation(x=mean_flipper_length, y=0, text=f"Mean: {mean_flipper_length:.2f}", showarrow=False, yshift=10, font=dict(color="blue"))
fig.add_annotation(x=median_flipper_length, y=0, text=f"Median: {median_flipper_length:.2f}", showarrow=False, yshift=25, font=dict(color="green"))
fig.add_annotation(x=mode_flipper_length, y=0, text=f"Mode: {mode_flipper_length:.2f}", showarrow=False, yshift=40, font=dict(color="red"))
fig.add_annotation(x=mean_flipper_length + std_flipper_length, y=0, text=f"1 SD: {std_flipper_length:.2f}", showarrow=False, yshift=55, font=dict(color="black"))

# Show the figure with annotations
fig.show(renderer="png")
```

```python
# Question 6 Figure
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
gifts_gender = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2024/2024-02-13/gifts_gender.csv')

# Calculate the spread (difference between max and min) for each gift type
spreads = gifts_gender.max(numeric_only=True) - gifts_gender.min(numeric_only=True)

# Plot the spreads
plt.figure(figsize=(10, 6))
spreads.plot(kind='bar')

# Add title and labels
plt.title('Spending Spread by Gift Type', fontsize=16)
plt.xlabel('Gift Type', fontsize=12)
plt.ylabel('Spread (Max - Min)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
