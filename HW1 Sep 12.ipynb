```python
import pandas as pd

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv"
df = pd.read_csv(url)

# Display the first few rows to inspect the data
print(df.head())

# Check for missing data counts in each column
missing_data_counts = df.isna().sum()
print("Missing data counts:\n", missing_data_counts)

rows, columns = df.shape 
# df.shape checks the size of the data set
#this fucntion is an attribute, whichb represent the static information that represent the obnject, in this case, the dataset
print(f"Total Rows: {rows}")
print(f"Total Columns: {columns}")


general_information = df.describe()
print("The general information of this data set is:\n", general_information )
```

```python
df.dropna()
   
#this is method in pandas that drops a ROW when there is one or more missing value 
```

```python
df.dropna(axis=0)
#Note:
# the same as dropping rows if theres one or more missing value, axis=0 means row
```

```python
df.dropna(axis=1)
#Note:
# dropping a COLUMN if theres one or more misssing value, axis = 1 means column
```

```python
df.dropna(how='all')
#Note
# it drops the rows / columns when all the values are missing.
# depends on what the condition of axis is at, axis=0 are rows, axis=1 are columns, when all the values are missing
```

```python
df.dropna(how='any')
#Note
# it drops any of the row or columns, depending on what axis, when theses rowsa or columns have one or more missing valus
# ans examplem or giving axies:
# df.dropna(how='any', axis=1)
```

```python
df.dropna(thresh=2)
#Note:
# keep any rows or columns have the corresponding numbers of value, 
# in this case, it kept any rows that has two (thresh = 2) or more certain values
```

```python
 df.dropna(subset=['name', 'gender'])
#Note
#only dropping rows with missing value with the specified columns
#in this case, it is name and gender
```
